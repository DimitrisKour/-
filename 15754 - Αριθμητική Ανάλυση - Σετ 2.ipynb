### Άσκηση 1:

Δίνεται το γραμμικό σύστημα,
$$
\begin{pmatrix} 2 & -3 & 1\\ 3 & -5 & 1\\ 0 & 2 & -3 \end{pmatrix}\cdot \begin{pmatrix} x\\y\\z \end{pmatrix} = 
\begin{pmatrix} 2 \\ 0 \\ 1 \end{pmatrix},
$$
το οποίο επιθυμούμε να λύσουμε με τη μέθοδο Gauss - Seidel και με αρχικές τιμές $(1, 1, 1)$. Η γενική αναδρομική σχέση γράφεται στη μορφή,

$$x_{i}^{(k+1)}=\frac{1}{a_{ii}}\left(b_{i}-\Sigma_{j=1}^{i-1}a_{ij}x_{j}^{(k+1)}-\Sigma_{j=i+1}^{N}a_{ij}x_{j}^{(k)} \right)$$

```python
# Δημιουργία προγρ΄αμματος που επιλύει οποιοδήποτε γραμμικό σύστημα με τη μέθοδο Gauss - Seidel με το σφάλμα που απαιτούμε.

def gauss_seidel(A, b, max_error, initial):
    N = len(A)
    iteration = []
    solution = []
    
    for n in range(N):
        solution.append([initial[n]])
    solution.append(iteration)
    
    for k in range(1000):
        iteration.append(k+1)
        
        for i in range(N):
            x = b[i]
            for j in range(N):
                if j != i:
                    if j < i:
                        x -= A[i][j]*solution[j][k+1]
                    else:
                        x -= A[i][j]*solution[j][k]
            solution[i].append(x / A[i][i])
    
        if (solution[0][k] - solution[0][k-1]) < max_error and (solution[1][k] - solution[1][k-1]) < max_error and \
        (solution[2][k] - solution[2][k-1]) < max_error:
            break
        
    return solution
```

```python
# Δεδομένα εκφώνησης.

initial = [1, 1, 1]

A = np.array([[2, -3, 1],
             [3, -5, 1],
             [0, 2, -3]])

b = np.array([2, 0, 1])
```

```python
# Ακρίβεια μέχρι 3 δεκαδικά ψηφία.

max_error = 10**(-3)

solution = gauss_seidel(A, b, max_error, initial)

n = len(solution)
N = len(solution[n - 1])

for i in range(N):
    print(str(solution[n - 1][i]) + ".\t x = %.3f"%solution[0][i] + "\t y = %.3f"%solution[1][i] + "\t z = %.3f"%solution[2][i])
```

```python
# Ακρίβεια μέχρι 6 δεκαδικά ψηφία.

max_error = 10**(-6)

solution = gauss_seidel(A, b, max_error, initial)

n = len(solution)
N = len(solution[n - 1])

for i in range(N):
    print(str(solution[n - 1][i]) + ".\t x = %.6f"%solution[0][i] + "\t y = %.6f"%solution[1][i] + "\t z = %.6f"%solution[2][i])
```

```python
# Ακρίβεια μέχρι 12 δεκαδικά ψηφία.

max_error = 10**(-12)

solution = gauss_seidel(A, b, max_error, initial)

n = len(solution)
N = len(solution[n - 1])

for i in range(N):
    print(str(solution[n - 1][i]) + ".\t x = %.12f"%solution[0][i] + "\t y = %.12f"%solution[1][i] + "\t z = %.12f"%solution[2][i])
```

### Άσκηση 2:

O πίνακας Hilbert $10 \times 10$ διαστάσεων αποτελείται από τα στοιχεία,
$$A_{ij}=\frac{1}{i+j-1}.$$
Ορίζουμε το διάνυσμα $\textbf x = [10, 9, 8, ..., 1]^{T}$, με στόχο να υπολογίσουμε το διάνυσμα $\textbf b=A\textbf x$. Ακολούθως, επιχειρούμε να επιλύσουμε το σύστημα,
$$A\textbf x_{num}=\textbf b$$
με τη μέθοδο του Gauss.

```python
# Ορισμός πίνακα Α και διανύσματος x.

N = 10

A = np.zeros((N, N))
x = np.array([])

for i in range(N): # Ορισμός Α.
    x = np.append(x, 10 - i)
    for j in range(N):
        A[i][j] =  1 / ((i+1) + (j+1) - 1) 
        # Θέτουμε i->i+1 και j->j+1 στη θεωρητική σχέση, αφού προγραμματιστικά ξεκινάμε από το 0 και όχι από το 1.
```

```python
# Ορισμός διανύσματος b.

b = np.array([])

def matrix_vector_product(A, x):
    N = len(A)
    b = np.zeros(N)
    
    for i in range(N):
        temp = 0
        
        for k in range(N):
            temp += A[i][k]*x[k]
            b[i] = temp
    
    return b

b = matrix_vector_product(A, x)
```

```python
# Εκτύπωση του πίνακα Α.

for i in range(N):
    for j in range(N):
        print("A[%d][%d] = %.6f"%((i+1), (j+1), A[i][j]))
    print("\n")
```

```python
# Εκτύπωση του διανύσματος b.

for i in range(N):
    print("b[%d] = %.6f"%((i+1), b[i]))
```

```python
# Σχηματισμός του επαυξημένου πίνακα Α:b.

def augmented_matrix(A, b):
    C = np.zeros((N, N+1))

    for i in range(N):
        for j in range(N+1):
            if j<N:
                C[i][j] = A[i][j]
            else:
                C[i][j] = b[i]
    
    return C

C = augmented_matrix(A, b)
```

```python
# Εκτύπωση του επαυξημένου πίνακα Α:b.

for i in range(N):
    for j in range(N+1):
        print("C[%d][%d] = %.6f"%((i+1), (j+1), C[i][j]))
    print("\n")
```

```python
# Δημιουργία συνάρτησης που εφαρμόζει τη μέθοδο του Gauss.

def Gauss(A, b):
    
    N = len(b)
    C = np.zeros((N, N+1))
    C = augmented_matrix(A, b)
  
    # Εύρεση μεγίστου.
    
    for i in range(0, N):    
        max_ = np.abs(C[i][i])  
        max_row = i 
        
        for k in range(i + 1, N):  
            if np.abs(C[k][i]) > max_:  
                max_ = np.abs(C[k][i])  
                max_row = k  
  
        # Εναλλαγή γραμμών.
    
        for k in range(i, N + 1):  
            temp = C[max_row][k]  
            C[max_row][k] = C[i][k]  
            C[i][k] = temp  
  
        # Μηδενισμός κατώτερων γραμμών.
    
        for k in range(i + 1, N):  
            c = - C[k][i] / C[i][i]  
            for j in range(i, N + 1):  
                if i == j:  
                    C[k][j] = 0  
                else:  
                    C[k][j] += c * C[i][j]  
  
    # Επίλυση της εξίσωσης Ux = b', εφόσον έχει σχηματιστεί ο άνω τριγωνικός πίνακας.
    
    x = np.zeros(N)
    
    for i in range(N - 1, -1, -1):  
        x[i] = C[i][N] / C[i][i]
        
        for j in range(i - 1, -1, -1):  
            C[j][N] -= C[j][i] * x[i]  
            
    return x 
```

```python
x_num = Gauss(A, b)

print("Προκύπτει αριθμητικά, μέσω της μεθόδου Gauss, το διάνυσμα x_num:")
for i in range(N):
    print("x_num[%d] = %.6f"%((i+1),x_num[i]))
```

Ορίζεται και η ακρίβεια της λύσης ως,
$$\alpha=\frac{\|x\| - \|x_{num}\|}{\|x\|},$$
η οποία υπολογίζεται ακολούθως.

```python
# Συνάρτηση για τον υπολογισμό του μέτρου ενός διανύσματος.

def norm(x):
    N = len(x)
    c = 0
    
    for i in range(N):
        c += x[i]**2
        
    c = np.sqrt(c)
    
    return c
```

```python
[norm(x), norm(x_num)]
```

```python
accuracy = (norm(x) - norm(x_num)) / norm(x)

print("H αριθμητική λύση υπολογίστηκε με ακρίβεια %.12f"%np.abs(accuracy) + ".")
```

### Άσκηση 3:

$α)$ Δίνεται ο πίνακας,
$$Α = \begin{pmatrix} 1 & 3 & 5 \\ 2 & 4 & 6 \\ 3 & 5 & 7
\end{pmatrix},$$
για τον οποίο αναζητούμε τη μέγιστη ιδιοτιμή, χρησιμοποιώντας τη μέθοδο των δυνάμεων, με 3 επαναλήψεις.
Η μεγαλύτερη ιδιοτιμή θα δίνεται από τη σχέση,

$$\lambda_{max}=\frac{x^{(k+1)}}{x^{(k)}}=\frac{\textbf A^{k+1}x}{\textbf A^{k}x},$$
όπου $\textbf x$ ένα τυχαίο διάνυσμα της επιλογής μας.

```python
# Δεδομένα εκφώνησης και τυχαίο διάνυσμα x.

A = np.array([[1, 3, 5],
             [2, 4, 6],
             [3, 5, 7]])

x = np.array([1, 1, 1])
```

```python
# Ορισμοί γινομένων πινάκων.

def matrix_product(A, B):
    N = len(A)
    C = np.zeros((N, N))
    
    for i in range(N):
        temp = 0
        for j in range(N):
            for k in range(N):
                temp += A[i][k]*B[k][j]
            C[i][j] = temp
    
    return C

def matrix_power(A, n):
    C = matrix_product(A, A)
    
    for i in range(n-2):
        C = matrix_product(C, A)
    
    return C
        

def matrix_vector_product(A, x):
    N = len(A)
    b = np.zeros(N)
    
    for i in range(N):
        temp = 0
        
        for k in range(N):
            temp += A[i][k]*x[k]
            b[i] = temp
    
    return b
```

Για τρείς επαναλήψεις, δηλαδή για $k=3$, η μεγαλύτερη ιδιοτιμή θα δίνεται προσεγγιστικά από τη σχέση,
$$\lambda_{max}=\frac{\textbf A^{4}\textbf x}{\textbf A^{3}\textbf x}.$$

```python
A3 = matrix_power(A, 3)
A4 = matrix_power(A, 4)
A3x = matrix_vector_product(A3, x)
A4x = matrix_vector_product(A4, x)
N = len(A3x)
solution = []

for i in range(N): # Η διαίρεση διανυσμάτων είναι συμβολική και αντιστοιχή σε διαίρεση των επιμέρους συντελεστών.
    solution.append(A4x[i]/A3x[i])

Lambda = np.average(solution) # Η ιδιοτιμή θα πρέπει να είναι πιο κοντά στη μέση τιμή.
```

```python
print("Μετά από 3 επαναλήψεις, η τιμή που λαμβάνουμε για τη μέγιστη ιδιοτιμή είναι %.12f"%Lambda +".")
```

$β)$ Εάν θέλουμε να βρούμε τη μικρότερη ιδιοτιμή, αρκεί να αξιοποιήσουμε το γεγονός ότι ο αντίστροφος πίνακας $A^{-1}$ έχει ιδιοτιμές $λ_{i}^{-1}$ ως προς τη βάση των ιδιοδιανυσμάτων $\textbf u^{(i)}$ του πίνακα $\textbf A$. Έτσι, αν ένα τυχαίο διάνυσμα $\textbf x$ είναι γραμμικός συνδυασμός των $\textbf u^{(i)}$, τότε, το γινόμενο $\textbf Α^{-1}\textbf x$ θα δίνει,
$$\textbf Α^{-1}\textbf x= \Sigma_{i=1}^{N}\left(\lambda_{i}^{-1}\right)\textbf u^{(i)} x,$$
από όπου, με όμοια λογική με αυτήν του προβλήματος της μέγιστης ιδιοτιμής, προκύπτει ότι,
$$r_{k}=\frac{\left(\textbf A^{-1}\right)^{k+1}\textbf x}{\left(\textbf A^{-1}\right)^{k}\textbf x}\approx \lambda_{min}^{-1},$$
ή, ισοδύναμα, 
$$\lambda_{min}\approx \frac{\left(\textbf A^{-1}\right)^{k}\textbf x}{\left(\textbf A^{-1}\right)^{k+1}\textbf x}.$$

Προκειμένου, όμως, ένας πίνακας $\textbf A$ να αντιστρέφεται, πρέπει η ορίζουσά του να είναι μη - μηδενική. 

Για έναν $3 \times 3$ πίνακα, η ορίζουσα δίνεται από τη σχέση,
$$det(\textbf A)=\epsilon_{ijk}\alpha_{1i}\alpha_{2j}\alpha_{3k}.$$

```python
# Κώδικας για ορίζουσα πίνακα 3 x 3.

def determinant(A):
    temp = A[0][0]*A[1][1]*A[2][2] + A[0][1]*A[1][2]*A[2][0] + A[0][2]*A[1][0]*A[2][1] - A[0][2]*A[1][1]*A[2][0] \
    - A[0][1]*A[1][0]*A[2][2] - A[0][0]*A[1][2]*A[2][1]
    
    return temp
```

```python
determinant(A)
```

O πίνακας $\textbf A$ έχει μηδενική ορίζουσα και, άρα, δεν μπορεί να αντιστραφεί. Συνεπώς, δεν μπορεί να υπολογιστεί η μικρότερη ιδιοτιμή με τη μέθοδο των δυνάμεων.

### Άσκηση 4:

Δίνεται η συνάρτηση,
$$f(x)=3xe^{x}-e^{2x}+3,$$
για την οποία έχουμε τις τιμές στα σημεία $x=-1,0,1,2.$

```python
# Δεδομένα εκφώνησης.

def f(x): return 3*x*np.exp(x) - np.exp(2*x) + 3

x_points_L = [-1, 0, 1, 2]
y_points_L = [f(i) for i in x_points_L]
```

$α)$ Θέλουμε να υπολογίσουμε το συμπτωτικό πολυώνυμο Lagrange που διέρχεται από αυτά τα σημεία. Ο τύπος του πολυωνύμου δίνεται από την,
$$P_{n}(x)=\Sigma_{i=0}^{n}f_{i}L_{i}(x),$$
όπου, $f_{i}=f(x_{i})$ και οι συντελεστές Lagrange δίνονται από τη σχέση,
$$L_{j}(x)=\frac{(x-x{1})(x-x_{2})...(x-x_{j-1})(x-x_{j+1})...(x-x_{n})}{(x_{j}-x_{1})(x_{j}-x_{2})...(x_{j}-x_{j-1})(x_{j}-x_{j+1})...(x_{j}-x_{n})}.$$

```python
# Συνάρτηση για τον υπολογισμό του πολυωνύμου Lagrange.

def Lj(x, j, x_points):
    n = len(x_points)
    a = 1 ; b = 1
    
    for i in range(n):
        if i != j:
            a *= (x - x_points[i])
            b *= (x_points[j] - x_points[i])
    
    return (a / b)

def Lagrange(x, x_points, y_points):
    n = len(x_points)
    temp = 0
    
    for i in range(n):
        temp += y_points[i]*Lj(x, i, x_points)
    
    return temp

# Lagrange(x, x_points_L, y_points_L): Η συνάρτηση που αντιστοιχεί στα δεδομένα του προβλήματος.
```

$β)$ Ακολούθως, υπολογίζουμε το εφαπτόμενο πολυώνυμο Hermite που περνά από τα δύο τελευταία σημεία $x=1,2$.

```python
# Δεδομένα εκφώνησης.

def Df(x): return (3*np.exp(x) + 3*x*np.exp(x) - 2*np.exp(2*x))

x_points_H = [1, 2]
y_points_H = [f(i) for i in x_points_H]
Dy_points_H = [Df(i) for i in x_points_H]
```

Το εφαπτόμενο πολυώνυμο Hermite αξιοποιεί και τη συμπεριφορά της παραγώγου των δεδομένων και οδηγεί στην τελική έκφραση,

$$P_{2n+1}(x)=\Sigma_{i=0}^{n}A_{i}(x)y_{i}+\Sigma_{i=0}^{n}B_{i}(x)y_{i}',$$

όπου,
$$A_{i}(x)=\left[1-2(x-x_{i})L_{i}'(x) \right]\cdot \left[L_{i}(x)\right]^{2}$$
$$B_{i}(x)=(x-x_{i})\cdot\left[L_{i}(x)\right]^{2}.$$

Εάν έχουμε δύο δεδομένα σημεία, η σχέση αυτή λαμβάνει τη μορφή,

$$P_{3}(x)=\Sigma_{i=0}^{1}A_{i}(x)y_{i}+\Sigma_{i=0}^{1}B_{i}(x)y_{i}',$$

ενώ σημειώνεται ότι τα πολυώνυμα Lagrange που χρησιμοποιούνται είναι τα,
$$L_{0}(x)=\frac{x-x_{1}}{x_{0}-x_{1}},$$
$$L_{1}(x)=\frac{x-x_{0}}{x_{1}-x_{0}},$$

με τις αντίστοιχες παραγώγους,

$$L_{0}'(x)=\frac{1}{x_{0}-x_{1}},$$
$$L_{1}'(x)=\frac{1}{x_{1}-x_{0}}.$$

```python
def L0(x, x_points): return Lj(x, 0, x_points)

def L1(x, x_points): return Lj(x, 1, x_points)


def DL0(x_points):
    return 1 / (x_points[0] - x_points[1])

def DL1(x_points):
    return 1 / (x_points[1] - x_points[0])

def Hermite(x, x_points, y_points, Dy_points):
    a = (1 - 2*(x - x_points[0])*DL0(x_points)) * (L0(x, x_points)**2) * y_points[0] # A0y0
    b = (1 - 2*(x - x_points[1])*DL1(x_points)) * (L1(x, x_points)**2) * y_points[1] # A1y1
    c = (x - x_points[0]) * (L0(x, x_points)**2) * Dy_points[0] # B0y0
    d = (x - x_points[1]) * (L1(x, x_points)**2) * Dy_points[1] # B1y1
    
    return a + b + c + d

# Hermite(x, x_points, y_points, Dy_points): Η συνάρτηση που αντιστοιχεί στα δεδομένα του προβλήματος.
```

```python
fig, ax = plt.subplots(1, figsize=(8, 6), tight_layout=True)

x1 = np.linspace(-1, 2, 1000)
x2 = np.linspace(1, 2, len(x1))

ax.scatter(x_points_L, y_points_L)
ax.plot(x1, f(x1), color = "black", linestyle = "dashed", label = r"$f(x)$")
ax.plot(x1, Lagrange(x1, x_points_L, y_points_L), label = r"$Lagrange$")
ax.plot(x2, Hermite(x2, x_points_H, y_points_H, Dy_points_H), label = r"$Hermite$")
ax.set_xlabel(r"$x$") ; ax.set_ylabel(r"$y$") ; ax.set_title(r"$Polynomial$ $Interpolation$ $Methods$.")
ax.legend(loc = "lower left") ; plt.rcParams["lines.linewidth"] = 0.8
ax.set_xlim(-1.05, 2.05)
plt.show()
```

Καταληκτικά, συγκρίνονται τα σχετικά σφάλματα της πραγματικής συνάρτησης $f(x)$ από τα πολυώνυμα $P^{L}_{n}(x)$ και $P^{H}_{2n+1}(x)$ που υπολογίστηκαν,
$$\epsilon_{L}=f(x)-P^{L}_{n}(x),$$
$$\epsilon_{H}=f(x)-P^{H}_{2n+1}(x).$$

```python
error_L = f(1.5) - Lagrange(1.5, x_points_L, y_points_L)
error_H = f(1.5) - Hermite(1.5, x_points_H, y_points_H, Dy_points_H)

print("Το σχετικό σφάλμα για το πολυώνυμο Lagrange είναι %.6f"%error_L + ".")
print("Το σχετικό σφάλμα για το πολυώνυμο Hermite είναι %.6f"%error_H + ".")
```

Είναι φανερό ότι το εφαπτόμενο πολυώνυμο Lagrange προσεγγίζει με μεγαλύτερη ακρίβεια τη συμπεριφορά της πραγματικής συνάρτησης $f(x)$, εφόσον αξιοποιεί τόσο τη συνέχεια των σημείων που γνωρίζουμε, αλλά και τις παραγώγους αυτών. Το συμπέρασμα αυτό διαπιστώνεται τόσο ποιοτικά μέσω του διαγράμματος που χαράχθηκε, αλλά και ποσοτικά από τα σχετικά σφάλματα που υπολογίστηκαν, αφού το σφάλμα $\epsilon_{H}$ είναι πολύ μικρότερο του σφάλματος $\epsilon_{L}.$
